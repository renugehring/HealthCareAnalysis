```python
# Read each year's file and concatenate
df_2019 = pd.read_csv('/content/drive/MyDrive/Data/All Beneficiary Years/beneficiary_2019.csv', sep="|")
df_2020 = pd.read_csv('/content/drive/MyDrive/Data/All Beneficiary Years/beneficiary_2020.csv', sep="|")
df_2021 = pd.read_csv('/content/drive/MyDrive/Data/All Beneficiary Years/beneficiary_2021.csv', sep="|")
df_2022 = pd.read_csv('/content/drive/MyDrive/Data/All Beneficiary Years/beneficiary_2022.csv', sep="|")
df_2023 = pd.read_csv('/content/drive/MyDrive/Data/All Beneficiary Years/beneficiary_2023.csv', sep="|")

# Concatenate all DataFrames
enrollment = pd.concat([df_2019, df_2020, df_2021, df_2022, df_2023], ignore_index=True)

```

```python
# prompt: are there any BENE_ID for whom any one of the 12 columns denoted by MDCR_STATUS_CODE_01 through MDCR_STATUS_CODE_12 equals 0

mdcr_status_cols = [f'MDCR_STATUS_CODE_{i:02d}' for i in range(1, 13)]


beneficiary_with_zero_status = enrollment[enrollment[mdcr_status_cols].isin([0]).any(axis=1)]['BENE_ID'].unique()
print(beneficiary_with_zero_status)
```

```python
# prompt: for each BENE_ID, is the average value across 12 columns denoted by MDCR_STATUS_CODE_01 through MDCR_STATUS_CODE_12 the same as the value in  MDCR_STATUS_CODE_01?  Are there exceptions to this?

# Calculate the average of MDCR_STATUS_CODE columns for each BENE_ID
enrollment['avg_mdcr_status'] = enrollment[mdcr_status_cols].mean(axis=1)

# Compare the average with MDCR_STATUS_CODE_01
enrollment['status_match'] = enrollment['avg_mdcr_status'] == enrollment['MDCR_STATUS_CODE_01']

# Find exceptions where the average and MDCR_STATUS_CODE_01 don't match
exceptions = enrollment[~enrollment['status_match']]
exceptions.head(5)


```

```python
# prompt: filter enrollment for the following columns BENE_ID
# STATE_CODE
# COUNTY_CD
# ZIP_CD
# BENE_BIRTH_DT
# SEX_IDENT_CD
# BENE_RACE_CD
# ESRD_IND
# BENE_DEATH_DT
# BENE_ENROLLMT_REF_YR

columns_to_keep = ['BENE_ID', 'STATE_CODE', 'COUNTY_CD', 'ZIP_CD', 'BENE_BIRTH_DT', 'SEX_IDENT_CD', 'BENE_RACE_CD', 'ESRD_IND', 'BENE_DEATH_DT', 'BENE_ENROLLMT_REF_YR']
enrollment = enrollment[columns_to_keep]
enrollment
```

```python
# prompt: what is the count of BENE_ID by BENE_ENROLLMENT_REF_YR
enrollment.groupby('BENE_ENROLLMT_REF_YR')['BENE_ID'].count()
```

```python
# prompt: are there duplicates by BENE_ID and BENE_ENROLLMENT_REF_YR

# Check for duplicates based on 'BENE_ID' and 'BENE_ENROLLMT_REF_YR'
duplicates = enrollment[enrollment.duplicated(subset=['BENE_ID', 'BENE_ENROLLMT_REF_YR'], keep=False)]

if not duplicates.empty:
  print("Duplicate entries found:")
  print(duplicates[['BENE_ID', 'BENE_ENROLLMT_REF_YR']])
else:
  print("No duplicate entries found based on BENE_ID and BENE_ENROLLMT_REF_YR.")
```

```python
# prompt: what is the maximum number of records by BENE_ID?

# Group by 'BENE_ID' and count occurrences
bene_id_counts = enrollment.groupby('BENE_ID').size()

# Find the maximum number of records for a single BENE_ID
max_records = bene_id_counts.max()

print(f"The maximum number of records for a single BENE_ID is: {max_records}")
```

```python
# prompt: Of BENE_IDS with 2023 in BENE_ENROLLMENT_REF_YR, how many are also in 2022, 2021, 2020, 2019?

import numpy as np
# Get the BENE_IDs present in 2023
bene_ids_2023 = enrollment[enrollment['BENE_ENROLLMT_REF_YR'] == 2023]['BENE_ID'].unique()

# Count how many of these BENE_IDs are also present in other years
counts = {}
for year in [2019, 2020, 2021, 2022]:
    bene_ids_year = enrollment[enrollment['BENE_ENROLLMT_REF_YR'] == year]['BENE_ID'].unique()
    count = np.sum(np.isin(bene_ids_2023, bene_ids_year))
    counts[year] = count
    print(f"Number of BENE_IDs in 2023 also present in {year}: {count}")
```

```python
# prompt: BENE_BIRTH_DT is read in 16-Aug-1999.  Read BENE_BIRTH_DT as datetime. Create AGE as the difference between year extracted from BENE_BIRTH_DATE and BENE_ENROLLMENT_REF_YR
enrollment['BENE_BIRTH_DT'] = pd.to_datetime(enrollment['BENE_BIRTH_DT'], format='%d-%b-%Y', errors='coerce')
# Extract the year from 'BENE_BIRTH_DT'
enrollment['BENE_BIRTH_YR'] = enrollment['BENE_BIRTH_DT'].dt.year

enrollment['AGE_ROUGH_CALC'] = enrollment['BENE_ENROLLMT_REF_YR'] - enrollment['BENE_BIRTH_YR']
```

```python
# prompt: create AGE_GRP from AGE_ROUGH_CALC. When AGE_ROUGH_CALC is equal to or below 18, AGE_GRP equals 'CHILD'.  Else if AGE_ROUGH_CALC is below 65, then AGE_GRP equals ADULT BELOW 65.  Else if AGE_ROUGH_CALC is equal to or above 65, then ADULT 65+.  Show me a frequency of AGE_GRP

import numpy as np
# Create AGE_GRP based on AGE_ROUGH_CALC
enrollment['AGE_GRP'] = np.where(enrollment['AGE_ROUGH_CALC'] <= 18, 'CHILD',
                               np.where(enrollment['AGE_ROUGH_CALC'] < 65, 'ADULT BELOW 65', 'ADULT 65+'))

# Show the frequency of AGE_GRP
print(enrollment['AGE_GRP'].value_counts())
```
